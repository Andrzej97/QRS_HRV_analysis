```python
import pandas as pd
import json
import matplotlib.pyplot as plt 
```

```python
results = pd.read_json('result.json')
```

```python
print(results.shape)
results.head()
```

```python
results_per_param = results.drop(columns=['filename']).groupby(['weight', 'threshold'])
results_per_param = results_per_param.sum().reset_index()
results_per_param['accuracy'] = results_per_param['true_positive']  / (results_per_param['true_positive'] + results_per_param['false_negative'] + results_per_param['false_positive']) 
results_per_param['precision'] = results_per_param['true_positive']  / (results_per_param['true_positive'] + results_per_param['false_positive'])
results_per_param['recall'] = results_per_param['true_positive']  / (results_per_param['true_positive'] + results_per_param['false_negative']) 
results_per_param['TP detection %'] = results_per_param['true_positive']  / results_per_param['annotation_count']
results_per_param['errors%'] = (results_per_param['false_positive'] + results_per_param['false_negative'])  / results_per_param['annotation_count']

weight_best_acc, threshold_best_acc, best_acc = results_per_param.sort_values('accuracy', ascending=False).iloc[0][['weight', 'threshold', 'accuracy']]

print(f'Best acc for weight: {weight_best_acc}, threshold: {threshold_best_acc} -> {best_acc}')
results_per_param
```

## Accuracy dla wszystkich ta≈õm dla (weight, threshold) 

```python
table_1 = pd.pivot_table(results_per_param, values='accuracy', index=['weight'], columns=['threshold'])
table_1
```

```python
weight_best_acc2, threshold_best_acc2, least_errors2 = results_per_param.sort_values('errors%', ascending=True).iloc[0][['weight', 'threshold', 'errors%']]

print(f'Least errors acc for weight: {weight_best_acc2}, threshold: {threshold_best_acc2} -> {least_errors2}')
```
