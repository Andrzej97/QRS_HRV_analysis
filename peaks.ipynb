```python
from multiprocessing.dummy import freeze_support
import wfdb
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
freeze_support()
# download_all_files()
record = wfdb.rdrecord('db/100', sampto=3000)
```

```python
print(record.p_signal)
print(len(record.p_signal))
wfdb.plot_items(signal=record.p_signal, title='MITDB Record 100')
```

```python
signal_ch0 = list(map(lambda x: x[0], record.p_signal))
signal_ch0
```

```python
plt.plot(signal_ch0)
```

```python
ecg = np.array(signal_ch0)
peaks, _ = find_peaks(ecg)
negPeaks, _ = find_peaks(-ecg)
```

```python
peaksX = peaks
peaksY = list(map(lambda x: ecg[x], peaks))

negPeaksX = negPeaks
negPeaksY = list(map(lambda x: ecg[x], negPeaks))

plt.plot(ecg)
plt.plot(peaksX, peaksY, 'ro')
plt.plot(negPeaksX, negPeaksY, 'bo')
plt.show()
```
